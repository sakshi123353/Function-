 1. What is the difference between a function and a method in Python?
      - 1. Function:
            A function is a block of code that performs a specific task. It is defined using the def keyword and can be called by its name.

           A function can be standalone and is not associated with any object or class.

          - Example:

           def add(a, b):
           
           return a + b
           
           result = add(2, 3)
           
           print(result)  # Output: 5
       - 2. Method:
            A method is a function that is associated with an object (usually an instance of a class) or a class itself.

            Methods are defined within a class and are typically called using the syntax object.method() or Class.method() for class methods.

            Methods often operate on data contained within the object (via self or cls), and they can modify the object's state.

            - Example:

            class MyClass:
            
            def __init__(self, value):
            
            self.value = value
    
           def show_value(self):
           
           print(self.value)

           obj = MyClass(10)
           
           obj.show_value()  # Output: 10

  - Function Parameters:

They allow the function to accept different inputs each time it is called.
  
  EXAMPLE :-

They are the real data you use when calling the function.
   EXAMPLE :-  
    - greet("Alice")  # "Alice" is an argument
greet("Bob")    # "Bob" is an argument


 3. What are the different ways to define and call a function in Python?
    - In Python, there are several ways to define and call functions. Here are the main approaches:

    - 1. Defining and Calling a Regular Function:
     Define: Using the def keyword to define the function.
     Call: Simply call the function by using its name.
     
      - example :-

     def greet(name):
            return f"Hello, {name}!"
       Call the function
      print(greet("Alice"))

     - 2. Using Lambda Functions (Anonymous Functions):
        Define: Lambda functions are small, anonymous functions created using the lambda keyword.
       Call: These are often used for short, one-liner operations.
     
          - example:-

         greet = lambda name: f"Hello, {name}!"

            Call the lambda function
           print(greet("Bob"))

      - 3. Using Function Arguments with Default Values:
      Define: You can set default values for function parameters.
       Call: The caller can provide arguments, but if they don't, default values are used.
        - example :-

       def greet(name="Guest"):
         return f"Hello, {name}!"

       Call with argument
      print(greet("Charlie"))

     Call without argument (uses default)
      print(greet())

     - 4. Using Variable-Length Arguments (*args and **kwargs):
         Define: The *args allows for an arbitrary number of positional arguments, and **kwargs allows for an arbitrary number of keyword arguments.
       Call: These allow flexibility in the number of arguments passed.
          - example

       def greet(*names):
       return " ".join(names)

       Call with multiple arguments
      print(greet("Alice", "Bob", "Charlie"))

      def display_info(**info):
      return f"Name: {info['name']}, Age: {info['age']}"

     Call with keyword arguments
     print(display_info(name="David", age=30))

   - 5. Nested Functions:
     Define: A function defined inside another function.
     Call: You can call the nested function from within the outer function.
        - exapmle

       def outer_function():
        def inner_function():
        return "This is the inner function!"
    
          return inner_function()

          Call the outer function
        print(outer_function())

  - 6. Function References:
      Define: Functions can be stored in variables and called via the variable name.
      Call: Using the function reference.

        - example

        def greet(name):
            return f"Hello, {name}!"

        func = greet
      Call using the variable reference
      print(func("Eve"))

     - 7. Calling Functions Using callable():
      Define: Any object that can be called is considered callable.
      Call: You can check if an object is callable using callable(), and then call it.
        - example
        
     def greet(name):
          return f"Hello, {name}!"

     func = greet
    if callable(func):
    print(func("Frank"))

 5. What are iterators in Python and how do they differ from iterables?
  - Iterator:
An iterator is an object that represents a stream of data and allows you to iterate over it. An iterator keeps track of its position during iteration.
An iterator must implement two methods:

       __iter__(): Returns the iterator object itself. This is required for the iterator to be used in an iteration context (like a for loop).

      __next__(): Returns the next item from the collection. When there are no more items to return, it raises the StopIteration exception.

 - Key Differences:

Iterable:

Is a container or collection of objects (like a list, string, or set).
Implements __iter__() method to return an iterator.

Iterator:

Is an object that represents a sequence of items and tracks the state of iteration.
Implements both __iter__() and __next__() methods.

Example:

# Example of an iterable

# Example of an iterator
print(next(iterable))  # Outputs 1

 7. What are the advantages of using generators over regular functions?
   - Generators are more memory-efficient and faster than regular functions because they produce values one at a time (lazy evaluation), rather than generating everything upfront. They allow for handling large datasets, infinite sequences, and can simplify iteration logic, while using less memory and offering better performance.

9. Explain the purpose and usage of the map() function in Python.
 - The map() function in Python applies a given function to each item in an iterable (like a list or tuple) and returns an iterator with the results.

Syntax:
map(function, iterable)

Example:
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x ** 2, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16]

map() is useful for applying a transformation to all elements in an iterable in a concise way.

```python
def square_numbers(numbers):
    return [num ** 2 for num in numbers]

# Example usage:
nums = [1, 2, 3, 4, 5]
squared_nums = square_numbers(nums)
print(squared_nums)  # Output: [1, 4, 9, 16, 25]
```

```python
def is_prime(n):
    """Check if a number is prime."""
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Checking prime numbers from 1 to 200
prime_numbers = [n for n in range(1, 201) if is_prime(n)]
print("Prime numbers from 1 to 200:", prime_numbers)
```

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration

        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

# Example usage:
n = 10
fib_iterator = FibonacciIterator(n)
for num in fib_iterator:
    print(num)
```

```python
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage:
for power in powers_of_two(5):
    print(power)

```

```python
def read_file_line_by_line(file_path):
    """Generator function to read a file line by line."""
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                yield line.strip()  # Remove leading/trailing whitespace
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
if __name__ == "__main__":
    file_name = "sample.txt"  # Replace with your file name

    # Writing sample data to file for demonstration
    with open(file_name, "w", encoding="utf-8") as f:
        f.write("Hello, World!\n")
        f.write("This is a sample file.\n")
        f.write("It has multiple lines.\n")

    # Reading file using generator function
    print("Reading file contents:")
    for line in read_file_line_by_line(file_name):
        print(line)
```

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 30, 37, 100]

# Use map() to convert to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Print results
print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
def remove_vowels(string):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, string))

# Example usage
input_string = input("Enter a string: ")
result = remove_vowels(input_string)
print("String without vowels:", result)
```
